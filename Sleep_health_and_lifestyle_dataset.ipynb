# Assignment1_Sleeping_Health_and_Lifestyle

```python
#Importing of Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Uploading the Dataset and viewing the Dataset

```python
#Uploading the Dataset
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

#Displaying of the data
df.head()
```

# Familiarization of the Data in the Dataset

```python
#Data familiarization

print(df.shape)  
print(df.info()) 
print(df.describe())
```

# Distribution of Dataset by Age,Occupation and Gender

```python
# Color customization
colors = ['black']
# Creating a histogram
plt.hist(df['Age'], color=colors)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Color customization
colors = ['green']
# Create a bar chart
df['Gender'].value_counts().plot(kind='bar', color=colors)
plt.title('Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('count')
plt.show()
```

```python
# Color customization
colors = ['gold']

# Creating a bar chart
df['Occupation'].value_counts().plot(kind='bar', color=colors)
plt.title('DISTRIBUTION OF OCCUPATIONS')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.show()
```

# Comparison between Sleep disorder and Gender

```python
grouped_df = df.groupby("Gender")["Sleep Disorder"].value_counts().unstack()
grouped_df.plot(kind='bar')
plt.title("Sleep Disorder Prevalence by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Sleep Disorder")
plt.show()
```

# Comparison of Sleep Duration with Qualityof Sleep

```python
# Create a scatter plot
plt.scatter(df['Sleep Duration'], df['Quality of Sleep'])
plt.title('Comparison Between Sleep Duration and Quality of Sleep')
plt.xlabel('Sleep Duration')
plt.ylabel('Quality of Sleep')
plt.grid(True)
plt.show()
```

# The relationship between Sleep disorder prevalence and factors like BMI and Age.

```python
# Define a color list (you can customize these colors)
colors = ['green', 'purple', 'red']

# Create a bar chart with custom colors
grouped_df.plot(kind='bar', color=colors)

# Increase font size for x and y labels and title
plt.title('Relationship Between Sleep Disorder and Age', fontsize=16)
plt.xlabel("Age", fontsize=14)
plt.ylabel("Count", fontsize=14)

# Increase font size for legend labels
plt.legend(title="Sleep Disorder", labels=grouped_df.columns, fontsize=12)

# Optional: Increase padding between axes and labels
plt.subplots_adjust(left=0, right=2) 

plt.xticks(rotation=45)  

plt.grid(axis='y') 

plt.show()
```

```python
#Creating a cross-tabulation table
cross_tab = pd.crosstab(df['Sleep Disorder'], df['BMI Category'])

#Display the cross tabulation table
print(cross_tab)

#Creating a heatmap from the cross tabulation
my_cmap = ['red','yellow','blue']
sns.heatmap(cross_tab, annot=True, cmap= my_cmap, fmt='d')

#Title and labels
plt.title("Sleep Disorder Prevalence by BMI Category")
plt.xlabel('BMI Category')
plt.ylabel('Sleep Disorder')

#Display the plot
plt.show()
```

# Relationship between Physical Activity and Heart Rate

```python
# Create a scatter plot
plt.scatter(df['Physical Activity Level'], df['Heart Rate'])
plt.title('Relationship Between Physical Activity and Heart Rate')
plt.xlabel('Physical Activity Level')
plt.ylabel('Heart Rate')
plt.grid(True)
plt.show()
```

# Distribution of Daily steps by BMI Category

```python
# Creating a histogram
df.hist(column='Daily Steps', by='BMI Category', bins=15, figsize=(18, 10))
plt.suptitle('Distribution of Daily Steps by BMI Category')
plt.show()
```

```python
# Creating a scatter plot
plt.scatter(df['Stress Level'], df['Quality of Sleep'])
plt.title('Relationship Between Stress Level and Quality of Sleep')
plt.xlabel('Stress Level')
plt.ylabel('Quality of Sleep')
plt.grid(True)
plt.show()
```
